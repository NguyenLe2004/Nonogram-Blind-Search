```python
from main import NonogramGame
import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
GRID_SIZE = 5
CELL_SIZE = 40

PUZZLE = [
    [" ", " ", "X", "X", "X"],
    [" ", " ", "X", " ", "X"],
    ["X", "X", " ", " ", " "],
    ["X", "X", " ", " ", " "],
    ["X", "X", "X", "X", " "]
]
```

```python
def gen_PUZZLE(size) : 
    return [[random.choice([" ", "X"]) for i in range(size)] for j in range(size)]
```

```python
game = NonogramGame(gen_PUZZLE(5))
time_elapsed,memMb = game.start_solving()
print((time_elapsed,memMb) )
```

```python
history = []
for size in range(3,16) :
    print(size , " " , end = " ")
    game = NonogramGame(gen_PUZZLE(size))
    time_elapsed,memMb = game.start_solving()
    history.append((time_elapsed,memMb))
```

```python
sns.pointplot(x=range(3,3+len(history)) , y = [h[0] for h in history])
plt.xlabel("Size")
plt.ylabel("Time(s)")
plt.show()
```

```python
sns.pointplot(x=range(3,3+len(history)) , y = [h[1] for h in history])
plt.xlabel("Size")
plt.ylabel("Memory (MB)")
plt.show()
```
